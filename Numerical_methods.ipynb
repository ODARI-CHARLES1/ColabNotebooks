```python
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
import numpy as np
```

#Interpolation of data.

```python

#Lagrange interpolation................. by charles odari
x=np.array([0,1,2,3]);
y=np.array([6,7,10,15])
poly=lagrange(x,y)

x=np.linspace(-3,3,100)
y=poly(x)

plt.plot(x,y)
plt.grid()
plt.show()


# y when x=4;
print(poly(1.4))
```

```python
x=np.array([0,0.3,0.6,0.8,1.1,1.2,1.71])
y=np.exp(x)
poly=lagrange(x,y)
x_range=np.linspace(0,5,100)
y_range=poly(x_range)

plt.plot(x_range,y_range)
plt.grid()
plt.show()

print(poly)
#y when x=0.5;
print(poly(0.5))
```

#  Numerical Differentiation Methods

Numerical methods can approximate the derivative of a function \( f(x) \) using discrete values. The most common methods are:

---

## 1. Forward Difference

The **forward difference** formula approximates the first derivative at point \( x \) as:

$$
f'(x) \approx \frac{f(x + \Delta x) - f(x)}{\Delta x}
$$

Where:
- \(Δx \) is a small increment in \( x \).


---

## 2. Backward Difference

The **backward difference** formula approximates the first derivative at point \( x \) as:

$$
f'(x) \approx \frac{f(x) - f(x - \Delta x)}{\Delta x}
$$

Where:
- (x-Δx) is the function value behind \( x \).

---

## 3. Central Difference

The **central difference** method gives a more accurate estimate:

$$
f'(x) \approx \frac{f(x + \Delta x) - f(x - \Delta x)}{2\Delta x}
$$

Where:
- It uses points on both sides of \( x \) for a symmetric approximation.

---

These methods are useful for approximating derivatives when analytic differentiation is not feasible, especially when working with discrete data or simulations.


```python
x=np.linspace(0,10,100)
y=np.sin(x)


#plot of the function sin(x)
plt.plot(x,y,"x")
plt.grid(True)
```

```python
#using the points to plot the derivative of the function sin(X)
dydx=np.zeros(len(x))
np.shape(dydx)
dx=x[1]-x[0]


#forward
dydx[0]=(y[1]-y[0])/dx

#central
for i in range(1,len(x)-1):
    dydx[i]=(y[i+1]-y[i-1])/(2*dx)

#backward
dydx[len(x)-1]=(y[len(x)-1]-y[len(x)-2])/dx


plt.plot(x,dydx)
plt.grid(True)
```
