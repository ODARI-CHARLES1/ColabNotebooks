Integrals and Derivatives


```python

x=np.linspace(1,10,100)
y=1/x**2+np.sin(x) #let us assume is an integral functions
plt.plot(x,y)
plt.grid()
```

```python
x=np.linspace(1,10,100)
y=-2/x**3+np.cos(x)
plt.plot(x,y)
plt.grid()
```

```python
N=10000
x=np.linspace(0,10,N+1)
y=np.exp(-x/10)*np.sin(x)
plt.plot(x,y,"r")
plt.grid()
```

```python
#mean value of values between 4and7
np.mean(y[(x>=4)*(x<=7)])
np.std(y[(x>=4)*(x<=7)])
```

```python
#find the 80 percentile
np.percentile(y[(x>=4)*(x<=7)],80)
```

```python

dydx=np.gradient(y,x)
dydx[1:]
plt.plot(x,dydx)
plt.grid()
```
